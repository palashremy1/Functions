1. What is the difference between a function and a method in Python?
  - A function is defined independently and can be called directly, while a method is a function associated with an object (e.g., list.append()).

2. Explain the concept of function arguments and parameters in Python.
  - Parameters are variables defined in a function, while arguments are the actual values passed when calling it.
  Eg: def add(a,b): return a+b → here a,b are parameters, add(3,5) → 3,5 are arguments.

3. What are the different ways to define and call a function in Python?
  - Functions are defined using def or lambda, and can be called by their name followed by parentheses.
  Eg: def greet(): print("Hello") → call with greet().

4. What is the purpose of the `return` statement in a Python function?
  - The return statement sends a value back from the function to the caller.

5. What are iterators in Python and how do they differ from iterables?
  - An iterable is an object you can loop over (like lists, strings), while an iterator is an object with __next__() method that fetches items one by one.
  Eg: iter([1,2,3]) gives an iterator.

6. Explain the concept of generators in Python and how they are defined.
  - Generators are special functions that yield values one by one using yield, instead of returning them all at once.

7. What are the advantages of using generators over regular functions?
  - They use less memory and allow lazy evaluation, producing values only when needed. Like iterating over large files line by line.

8. What is a lambda function in Python and when is it typically used?
  - A lambda is a small anonymous function defined with the lambda keyword.
  Eg: lambda x: x*x.

9. Explain the purpose and usage of the `map()` function in Python.
  - map() applies a function to every item of an iterable and returns an iterator.
  Eg: map(str.upper, ["a","b"]).

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - map() applies a function to each item.
  - filter() keeps only items that match a condition.
  - reduce() combines items into a single value.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even(numbers):
    total = 0                       # Start with total = 0
    for n in numbers:               # Loop through each number in the list
        if n % 2 == 0:              # Check if the number is even
            total += n              # Add even number to total
    return total

print(sum_even([1,2,3,4,5,6]))

```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n < 2:                        # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):   # Check divisibility up to sqrt(n)
        if n % i == 0:               # If divisible, not prime
            return False
    return True                      # Otherwise prime


# Print primes between 1 and 200
for num in range(1, 201):
    if is_prime(num):
        print(num, end=" ")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

# Iterator class for Fibonacci sequence
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms   # total number of terms to generate
        self.a, self.b = 0, 1
        self.count = 0       # to keep track of generated numbers

    def __iter__(self):
        return self          # the object itself is the iterator

    def __next__(self):
        if self.count < self.terms:     # stop when we reach the limit
            value = self.a              # current Fibonacci number
            self.a, self.b = self.b, self.a + self.b  # move forward
            self.count += 1
            return value
        else:
            raise StopIteration         # end of sequence


# Example usage
fib = Fibonacci(10)   # Generate 10 terms
for num in fib:
    print(num, end=" ")

```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(limit):
    for i in range(limit+1):
        yield 2**i

print(list(powers_of_two(5)))
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file(filename):
    with open(filename, "r") as f:
        for line in f:
            yield line.strip()
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

tuples = [(1, 5), (2, 1), (3, 3)]
sorted_list = sorted(tuples, key=lambda x: x[1])
print(sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

temps_c = [0, 20, 37, 100]
temps_f = list(map(lambda c: (c*9/5)+32, temps_c))
print(temps_f)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

s = "palash rammaiya"
result = "".join(filter(lambda ch: ch.lower() not in "aeiou", s))
print(result)
```
